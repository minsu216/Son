데이터 전처리 과정(value, wage 단위 변경)

```python
def convert_value_wage(value_or_wage):
    if 'M' in value_or_wage:
        return int(float(value_or_wage.replace('M', ''))*1000000)
    elif 'K' in value_or_wage:
        return int(float(value_or_wage.replace('K', ''))*1000)
    else:
        return int(float(value_or_wage))

# 'value'와 'wage' 열에 함수 적용
```

```python
df['value'] = df['value'].apply(convert_value_wage)
df['wage'] = df['wage'].apply(convert_value_wage)
```
